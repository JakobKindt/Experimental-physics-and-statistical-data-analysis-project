```python
import numpy as np
import matplotlib.pyplot as plt
mus = []
sigmas = []

I1 = np.array([1.485, 1.521, 1.477, 1.502, 1.563, 1.588, 1.547, 1.531, 1.531, 1.533])*10**-3
I2 = np.array([2.174, 2.225, 2.212, 2.191, 2.242, 2.293, 2.207, 2.259, 2.229, 2.269])*10**-3
I3 = np.array([3.298, 3.287, 3.323, 3.334, 3.280, 3.343, 3.327, 3.303, 3.260, 3.287])*10**-3
I4 = np.array([4.529, 4.561, 4.506, 4.526, 4.534, 4.606, 4.571, 4.611, 4.531])*10**-3

Is = [I1, I2, I3, I4]

for I in Is:
    mu = np.mean(I)
    sigma = np.std(I, ddof = 1) # Division by N-1 instead of N
    mus.append(mu)
    sigmas.append(sigma)
mus = np.array(mus)
sigmas = np.array(sigmas)
print('µ: ', mus)
print('σ: ', sigmas)
print('σ/µ: ', sigmas/mus)

fig, ax = plt.subplots()
ax.set_title('$\sigma$ as function of µ')
ax.plot(mus, sigmas, marker = 'o')
ax.set_xlabel('µ')
ax.set_ylabel('$\sigma$')
plt.show()

fig, ax = plt.subplots()
ax.set_title('$\sigma$/µ as function of µ')
ax.plot(mus, sigmas/mus, marker = 'o')
ax.set_xlabel('µ')
ax.set_ylabel('$\sigma$/µ')
plt.show()
```
